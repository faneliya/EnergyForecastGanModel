```python
from bs4 import BeautifulSoup
```

```python
html = driver.page_source
```

```python
soup = BeautifulSoup(html,'html5lib')
```

```python
title = soup.find_all('a',{'aria-label'})
```

```python
title
```

```python
title_list=[]
for i in title:
    d=i.get_text()
    title_list.append(d)
```

```python
title_list
```

```python
date=soup.find_all('span')
```

```python
date
```

```python
date_list=[]
for i in date:
    d=i.get_text()
    date_list.append(d)
```

```python
date_list
```

```python
df_show_info = pd.DataFrame(
    {'Article Title': title_list})
```

```python
df_show_info.to_csv('titletest.csv')
```
